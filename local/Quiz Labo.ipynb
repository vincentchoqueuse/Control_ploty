```python
import sys
sys.path.append("../")

from ENIB_control import *
from control import tf, c2d, feedback
import numpy as np
import plotly
import json
from lib import Question,Quiz, Category
```

```python
Te=4
def Cz(K,Ti,Te=4):
    return K+K*tf([Te/Ti,0],[1,-1],Te)

def draw_sys(Fz):
    fig = figure("nichols")
    fig.plot(Fz,w=np.logspace(-3,0,200))
    fig.grid(show_phase=False)
    return fig
```

```python
class General_Question(Question):
    
    def get_template_name(self):
        return '{}.xml'.format(self.name)
    
    def __init__(self,data,answer,name = "temp"):
        self.data = data
        self.answer = answer
        self.object = None
        self.name = name
    
    def get_context_data(self):
        context = super().get_context_data()
        context["data"] = self.data
        context["answer"] = self.answer
        return context
    
name = "DÃ©faut pour ASN S6P/eval_labo3/identifications"
quiz = Quiz(name = name)

category1 = Category("{}/nichols".format(name))
category2 = Category("{}/step".format(name))
category3 = Category("{}/rlocus".format(name))
#quiz.add(category)
```

```python
sys1 = tf([2.3],[600,70,1])
sys2 = tf([1],[500,30,1])
sys3 = tf([2],[300,50,1])
sys4 = tf([5],[100,70,1])

Fz_list = []
for sys in [sys1,sys2,sys3,sys4]:
    Fz = c2d(sys,Te)
    Fz_list.append(Fz)
    
w=np.logspace(-3,0,500)
```

# Systeme 1

```python
wr = 0.101
Ti = 10/wr
C =Cz(2.2,Ti)
H_list =[Fz_list[0],4*Fz_list[0],C*Fz_list[0]]
```

```python
fig = figure("nichols")
fig.plot(H_list[0],w=w,label="sys1")
fig.plot(H_list[1],w=w,label="sys2")
fig.plot(H_list[2],w=w,label="sys3")
fig.grid(cm = np.array([6,3,1,0,-0.5,-0.9,-1,-3,-6]),show_phase=False)
fig.xlim([-250,0])
fig.ylim([-30,40])

category1.add(General_Question(fig.json(),[7.2,4,-0.9,3.9,0,1,0.068,53,16.5],name="nichols"))
fig.show()
```

```python
fig = figure("time") 
fig.plot(feedback(H_list[1],1))
fig.xlim([0,250])
category2.add(General_Question(fig.json(),[0.25,0.901,0.32],name="step"))
fig.show()
```

```python
fig = figure("rlocus") 
fig.plot(H_list[0])
category3.add(General_Question(fig.json(),[7.80,4.763],name="rlocus"))
fig.show()
```
